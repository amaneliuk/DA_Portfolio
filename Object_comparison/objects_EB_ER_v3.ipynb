## ТЗ
- Сравнить объекты в выгрузке из ЭБ с объектами в ЭР. Проверить, все ли объекты ЭР соответствуют объектам в ЭБ, в привязке к мероприятиям ЭР (результатам ФП в ЭБ).
- Сколько объектов и какого типа есть в ЭР сверх ЭБ

<i>*ОКС - объект капитального строительства</i>

```python
#Иморт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Загружаемые файлы
vygr_eb = 'obj_EB_1211.xlsx' # - выгрузка из ГИИС "Электронный бюджет" за 12.11.2021
vygr_nsi_4856 = 'export_4856_1211.csv' # - выгрузка каталога 4856 (объекты) из НСИ за 12.11.2021
vygr_nsi_400 = 'export_400_1211.csv' # - выгрузка каталога 400 (мероприятия) из НСИ за 12.11.2021
er_976 = '976_lite.xlsx' # - файл с 976 мероприятиями Эталонного реестра
```

```python
#Выгрузка из ЭБ по объектам в привязке с результатам ФП
eb_objects = pd.read_excel(vygr_eb)
eb_objects.head(3)
```

```python
#Загрузка Эталонного реестра мероприятий 976
er_units_976 = pd.read_excel(er_976, sheet_name = '976')
er_units_976.head(3)
```

```python
er_units_976.info()
```

```python
#Список id 976 мероприятий Эталонного реестра из НСИ
list_976 = er_units_976['global_id'].to_list()
```

```python
#Загрузка  реестра мероприятий из НСИ за 12.11.2021
nsi_units_1211 = pd.read_csv(vygr_nsi_400, delimiter = ';', encoding = 'ANSI', skiprows = 1)
nsi_units_1211.head(3)
```

```python
#Количество мероприятиц в выгрузке НСИ
len(nsi_units_1211)
```

```python
#Функция определения состоит ли id мероприятия в Эталонном реестре
def id_in_er(data):
    if data in list_976:
        data = 1
    else:
        data = 0
    return data
```

```python
#Применение функции и создание нового столбца
nsi_units_1211['ЭР'] = nsi_units_1211['global_id'].apply(id_in_er)
```

```python
#Отфильтруем из выгрузки НСИ за 12.11.2021 мероприятия, содержащиеся в Эталонном реестре
nsi_er_units_1211 = nsi_units_1211[nsi_units_1211['ЭР'] == 1]
len(nsi_er_units_1211)
```

<div class="alert alert-success">
Таким образом получаем, что в НСИ по состоянию на 12.11.2021 загружено <b>972</b> мероприятия Эталонного реестра (из 976).
    </div>

```python
nsi_er_units_1211.head(3)
```

```python
#Создадим сокращенную таблицу с нужными полями
clean_nsi_er_units_1211 = nsi_er_units_1211[['global_id', 'system_object_id', 'Наименование мероприятия (результата)', \
                                             'Тип мероприятия (результата)', \
                                             'Код из Электронного бюджета']]
clean_nsi_er_units_1211
```

```python
#Загрузка  реестра объектов (каталог 4856) из НСИ за 12.11.2021
nsi_objects_1211 = pd.read_csv(vygr_nsi_4856, delimiter = ';', encoding = 'ANSI', skiprows = 1)
nsi_objects_1211.head(3)
```

```python
#Отфильтруем из выгрузки объектов НСИ за 12.11.2021 объекты, привязанные к мероприятия, содержащиеся в Эталонном реестре
nsi_objects_1211['ЭР'] = nsi_objects_1211['global_object_id'].apply(id_in_er)
nsi_er_objects_1211 = nsi_objects_1211[nsi_objects_1211['ЭР'] == 1]
nsi_er_objects_1211.head(3)
```

```python
nsi_er_objects_1211.info()
```

```python
print('Количество объектов в выгрузке НСИ за 12.11.2021 привязанных к мероприятиям ЭР:', len(nsi_er_objects_1211))
```

```python
#Группировка объектов по мероприятиям
nsi_objects_by_units = nsi_er_objects_1211.groupby('global_object_id')['global_id'].count()\
    .to_frame()\
    .reset_index()
nsi_objects_by_units.rename(columns={'global_id':'Кол-во объектов'}, inplace=True)
nsi_objects_by_units
```

```python
#Проверка количества объектов
nsi_objects_by_units['Кол-во объектов'].sum()
```

```python
#Проверка количества уникальных id мероприятий
nsi_er_objects_1211['global_object_id'].nunique()
```

```python
#Соединим таблицы чтобы получить реквизиты мероприятия
nsi_objects_by_units = clean_nsi_er_units_1211.merge(nsi_objects_by_units, how = 'left'\
                                                     , left_on = 'global_id', right_on = 'global_object_id')
nsi_objects_by_units
```

```python
#Загрузка справочника типов мероприятий
spr_unit_type = pd.read_excel('spr_unit_type.xlsx')
spr_unit_type
```

```python
#Подгрузим справочник типов мероприятий
nsi_objects_by_units = nsi_objects_by_units.merge(spr_unit_type, how = 'left', left_on = 'Тип мероприятия (результата)', \
                           right_on = 'Идентификатор')
nsi_objects_by_units
```

```python
#Посчитаем кол-во уникальных идентификаторов значения результата ФП
eb_objects_by_results['Идентификатор значения результата ФП'].nunique()
```

```python
#Сформируем списки id результата ФП и id значения результата ФП
id_results_eb = eb_objects_by_results['Идентификатор результата ФП'].unique().tolist()
id_values_results_eb = eb_objects_by_results['Идентификатор значения результата ФП'].unique().tolist()
```

```python
def id_in_spisok(data):
    if data in spisok:
        data = 1
    else:
        data = 0
    return data
```

```python
#Проверим есть ли совпадения в мероприятиях НСИ с ЭБ по id "Идентификатор результата ФП"
spisok = id_results_eb
nsi_objects_by_units['match'] = nsi_objects_by_units['system_object_id'].apply(id_in_spisok)
nsi_objects_by_units[nsi_objects_by_units['match'] == 1]
```

```python
nsi_objects_by_units['match'] = nsi_objects_by_units['Код из Электронного бюджета'].apply(id_in_spisok)
nsi_objects_by_units[nsi_objects_by_units['match'] == 1]
```

<div class="alert alert-success">
Таким образом, мы выяснили что "Идентификатор результата ФП" не присутствует в полях выгрузки из НСИ.
</div>

```python
#Проверим есть ли совпадения в мероприятиях НСИ с ЭБ по id "Идентификатор значения результата ФП"
spisok = id_values_results_eb
nsi_objects_by_units['match'] = nsi_objects_by_units['system_object_id'].apply(id_in_spisok)
nsi_objects_by_units[nsi_objects_by_units['match'] == 1]
```

```python
nsi_objects_by_units['match'] = nsi_objects_by_units['Код из Электронного бюджета'].apply(id_in_spisok)
nsi_objects_by_units[nsi_objects_by_units['match'] == 1]
```

<div class="alert alert-success">
Также и "Идентификатор значения результата" не находится в выгрузке НСИ.
</div>

```python
#Проверка функции
spisok = id_values_results_eb
eb_objects_by_results['match'] = eb_objects_by_results['Идентификатор значения результата ФП'].apply(id_in_spisok)
len(eb_objects_by_results[eb_objects_by_results['match'] == 1])
```

<div class="alert alert-success">
Функция работает корректно. В выгрузке из ЭБ все находит.
</div>

```python
nsi_objects_by_units.head(3)
```

```python
#Cгруппируем кол-во объектов в привязке к результатам ФП в выгрузке из ЭБ
merged_eb_objects_by_results = eb_objects_by_results.groupby\
    ('Наименование результата')['Кол-во ОКС/объектов'].sum().to_frame().reset_index()
merged_eb_objects_by_results
```

```python
merged_eb_objects_by_results['Кол-во ОКС/объектов'].sum()
```

<div class="alert alert-success">
В выгрузке из ГИИС "Электронный бюджет" <b>3523</b> объекта привязаны к <b>132</b> результатам федеральных проектов.

<div class="alert alert-success">
Так как мы выяснили, что ни по каким id у нас совпадений в выгрузках нет, придется соединять полученные таблицы по по наименованию мероприятия (результата).
</div>

```python
#Соединим выгрузки НСИ и ЭБ (по названию мероприятия)
match = nsi_objects_by_units.merge(merged_eb_objects_by_results, how = 'inner', \
                                   left_on = 'Наименование мероприятия (результата)',\
                           right_on = 'Наименование результата')
match
```

```python
#Проверка на дубликаты
match['global_id'].nunique()
```

```python
#Длина таблицы
len(match)
```

```python
#Приведем таблицу с результатом к удобному виду
match.drop(['system_object_id', 'global_object_id', 'match', 'Наименование результата'], axis = 1, inplace = True)

match = match[['global_id', 'Код из Электронного бюджета', 'Наименование мероприятия (результата)', 'Кол-во объектов'\
      , 'Идентификатор', 'Наименование', 'Кол-во ОКС/объектов']].copy()

match.rename(columns = {'Кол-во объектов':'Кол-во_объектов_ЭР','Идентификатор':'Тип мероприятия', \
                        'Наименование':'Наименование типа мероприятия', 'Кол-во ОКС/объектов':'Кол-во_объектов_ЭБ'},\
             inplace = True)

match['Кол-во_объектов_ЭР'].fillna(value = 0, inplace = True)

match['Кол-во_объектов_ЭР'] = match['Кол-во_объектов_ЭР'].astype('int32')
```

```python
#Добавим столбец с разницей
match['Разница ЭР - ЭБ'] = match['Кол-во_объектов_ЭР'] - match['Кол-во_объектов_ЭБ']
```
