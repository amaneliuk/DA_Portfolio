```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('input.xlsx')
df
```

```python
df_3 = df.copy()
```

```python
#Функция вычисления среднего за 2 месяца
def ss_2_month(data_month_1, data_month_2):
    ss_2 = (data_month_1 + data_month_2) / 2
    return ss_2
```

```python
df['октябрь_сс_2'] = ss_2_month(df['август'], df['сентябрь'])
df['ноябрь_сс_2'] = ss_2_month(df['сентябрь'], df['октябрь_сс_2'])
df['декабрь_сс_2'] = ss_2_month(df['октябрь_сс_2'], df['ноябрь_сс_2'])
```

```python
#Функция вычисления среднего за 3 месяца
def ss_3_month(data_month_1, data_month_2, data_month_3):
    ss_3 = (data_month_1 + data_month_2 + data_month_3) / 3
    return ss_3
```

```python
df_3['октябрь_сс_3'] = ss_3_month(df_3['июль'], df_3['август'], df_3['сентябрь'])
df_3['ноябрь_сс_3'] = ss_3_month(df_3['август'], df_3['сентябрь'], df_3['октябрь_сс_3'])
df_3['декабрь_сс_3'] = ss_3_month(df_3['сентябрь'], df_3['октябрь_сс_3'], df_3['ноябрь_сс_3'])
```
