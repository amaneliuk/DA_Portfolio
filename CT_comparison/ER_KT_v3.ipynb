```python
#Проверка на дубли в наименованиях КТ в одном мероприятии
test_df = df_4852_2021.groupby('global_object_id')[['Наименование КТ', 'Плановая дата КТ']].count()
test_df[test_df['Наименование КТ'] != test_df['Плановая дата КТ']]
```

Дублей не выявлено

```python
#Сводная таблица
ct_4852_count = df_4852_2021.groupby('global_object_id')['Наименование КТ'].count()\
    .to_frame()\
    .reset_index()

ct_4852_count.columns = ['id', 'Кол-во КТ']
```

```python
ct_4852_count
```

```python
er_fp.head(1)
```

```python
#Совмещение таблицы мероприятий с таблицей кол-ва КТ
ct_er_fp = er_fp.merge(ct_4852_count, how = 'left', left_on = 'global_id', right_on = 'id')
```

```python
ct_er_fp.head(3)
```

```python
ct_er_fp.info()
```

```python
#Оставление нужных столбцов в таблице
ct_er_short = ct_er_fp[['global_id', 
        'Наименование мероприятия (результата)',
        'Федеральный проект',
        'Код из Электронного бюджета',
         'id',
         'Кол-во КТ']]
ct_er_short
```

```python
#Суммарное кол-во КТ
ct_er_short['Кол-во КТ'].sum()
```

```python
#Загрузка выгрузки Минфина
mf_eb = pd.read_excel('ct_05102021_mf.xlsx')
```

```python
mf_eb.head(3)
```

```python
mf_eb.info()
```

```python
#Фильтрация только КТ
mf_eb_ct = mf_eb[mf_eb['Тип (результат, КТ)'] == 'КТ'].copy()
```

```python
mf_eb_ct.head(3)
```

```python
mf_eb_ct.info()
```

```python
print(mf_eb_ct['Идентификатор КТ'].count(), '- Всего')
print(mf_eb_ct['Идентификатор КТ'].nunique(), '- Уникальных')
```

```python
#Сортировка с годом
mf_eb_ct['Ct_year'] = mf_eb_ct['Плановый итоговый срок реализации'].dt.year
mf_eb_ct_2021 = mf_eb_ct[mf_eb_ct['Ct_year'] == 2021].copy()
```

```python
mf_result_ct = mf_eb_ct_2021.groupby('Идентификатор результата')['Тип (результат, КТ)'].count()\
    .to_frame()\
    .reset_index()
mf_result_ct.columns =['Идентификатор результата', 'Кол-во_КТ_ЭБ']
mf_result_ct
```

```python
merged_er_mf_ct = ct_er_short\
    .merge(mf_result_ct, left_on = 'Код из Электронного бюджета', right_on = 'Идентификатор результата', how = 'left')
merged_er_mf_ct.head()
```

```python
mf_eb.head(3)
```

```python
clean_res = merged_er_mf_ct.drop(['id', 'Идентификатор результата'], axis = 1)
clean_res.rename(columns={'Кол-во КТ':'Кол-во КТ_ЭР'}, inplace=True)
clean_res['Разница ЭР - ЭБ'] = clean_res['Кол-во КТ_ЭР'] - clean_res['Кол-во_КТ_ЭБ']
clean_res
```
